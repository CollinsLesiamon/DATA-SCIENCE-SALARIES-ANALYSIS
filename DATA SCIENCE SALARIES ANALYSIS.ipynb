## Importing Libraries

```python
import pandas as pd #for handling data and data structures.
import matplotlib.pyplot as plt #for plotting and data visualisation.
import plotly.express as px  # Used for interactive charts and data visualization
import seaborn as sns #for further enhanced visualisation.
```

## Loading The Dataset

```python
df = pd.read_csv("DataScience_salaries_2024.csv")
df.head()
```

## Initial Data Exploration

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['salary_in_usd'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Salary (USD)')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(df, x='work_year', y='salary_in_usd')
plt.title('Salary Distribution Over Time')
plt.xlabel('Year')
plt.ylabel('Salary (USD)')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(df, x="experience_level", y='salary_in_usd')
plt.xlabel("Experience Level")
plt.ylabel("Salary (USD)")
plt.title("Salary by Experience Level")
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(df, x="employment_type", y='salary_in_usd')
plt.xlabel("employment_type")
plt.ylabel("Salary (USD)")
plt.title("Salary by employment_type")
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(df, x="company_size", y='salary_in_usd')
plt.xlabel("company_size")
plt.ylabel("Salary (USD)")
plt.title("Salary by company_size")
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(df, x="remote_ratio", y='salary_in_usd')
plt.xlabel("remote_ratio")
plt.ylabel("Salary (USD)")
plt.title("Salary by remote_ratio")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(df, x="job_title", y="salary_in_usd")
plt.title('Distribution of Salary (USD)')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.show()
```
